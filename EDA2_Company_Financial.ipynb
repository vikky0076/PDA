**Advanced EDA & Business Insights**

Objective

Create new metrics and extract insights.

**Profit Margin**

```python
df['Profit_Margin'] = (df['Net_Profit_INR_Crores'] /
                       df['Revenue_INR_Crores']) * 100
```

**Revenue Growth Rate**

```python
df['Revenue_Growth'] = df['Revenue_INR_Crores'].pct_change() * 100
```

**Profit Margin Trend**

```python
sns.lineplot(x='Year', y='Profit_Margin', data=df)

plt.title("Profit Margin Trend")
plt.show()
```

**Revenue Growth Visualization**

```python
sns.barplot(x='Year', y='Revenue_Growth', data=df)

plt.title("Revenue Growth Rate")
plt.show()
```

**Financial Stability Ratio**

```python
df['Debt_to_Revenue'] = df['Debt_INR_Crores'] / df['Revenue_INR_Crores']
```

**Visualization**

```python
sns.lineplot(x='Year', y='Debt_to_Revenue', data=df)

plt.title("Debt to Revenue Ratio")
plt.show()
```
